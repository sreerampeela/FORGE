```python
import pandas as pd
import numpy as np
```

```python
exp = pd.read_csv('/home/nilabjab/cancer_dependency_project_nilabja/cancer_dependency_project/Datasets/DepMap/OmicsExpressionRNASeQCGeneCountProfile.csv', 
                  header = 0, index_col = 0)
exp.head()
```

```python
depmap_metadata = pd.read_csv('/home/nilabjab/cancer_dependency_project_nilabja/cancer_dependency_project/Datasets/DepMap/Model.csv', 
                              header = 0, index_col = 0)
depmap_metadata.head()
```

```python
depmap_metadata['ModelID2'] = depmap_metadata.index
```

```python
depmap_metadata.shape, exp.shape
```

```python
# profile data
profile_data = pd.read_csv('/home/nilabjab/cancer_dependency_project_nilabja/cancer_dependency_project/Datasets/DepMap/OmicsProfiles.csv', 
                           header = 0, index_col = 0)
profile_data.head()
```

```python
profile_data.reset_index(inplace = True)
```

```python
# merge three dfs
merged_df = pd.merge(
    left=profile_data,
    right=depmap_metadata,
    left_on='ModelID',
    right_on='ModelID2',
    how='inner'
).reset_index(drop=False)
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df.shape
```

```python
common_pt_ids = np.intersect1d(merged_df['ProfileID'], exp.index)
len(common_pt_ids)
```

```python
exp_filtered = exp.loc[common_pt_ids]
exp_filtered.shape
```

```python
map_dict = dict(zip(merged_df['ProfileID'], merged_df['StrippedCellLineName']))
```

```python
map_dict2 = dict(zip(merged_df['ModelID'], merged_df['StrippedCellLineName']))
```

```python
exp_filtered['cell_line'] = exp_filtered.index.map(map_dict)
```

```python
exp_filtered.head()
```

```python
exp_filtered = exp_filtered.set_index('cell_line', drop = True)
```

```python
# Step 1: Drop all columns that start with "ERCC-"
exp_clean = exp_filtered.loc[:, ~exp_filtered.columns.str.startswith("ERCC-")]

# Step 2: Remove Ensembl IDs from remaining gene column names (keep only gene symbol)
exp_clean.columns = exp_clean.columns.str.replace(r"\s+\(ENSG\d+\)", "", regex=True)

exp_clean.shape
```

```python
with open('../Data/gene_names.txt', 'r') as f:
    gene_ids = [i.rstrip().strip() for i in f.readlines()]
    
len(gene_ids)
```

```python
len(np.intersect1d(gene_ids, exp_clean.columns))
```

```python
exp_clean = exp_clean[np.intersect1d(gene_ids, exp_clean.columns)]
exp_clean.head()
```

```python
exp_clean = exp_clean.groupby(exp_clean.columns, axis=1).sum()
```

```python
exp_clean.shape
```

```python
exp_clean.to_csv('../Data/cleaned_depMap_rawExp.csv', index=True)
```
